```python
import librosa
import numpy as np
import IPython.display as ipd
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import sklearn

path_of_audio = 'metal/metal.00073.au.wav'
the_way = "testsongs/example1.wav"

x , sr = librosa.load(path_of_audio, sr=None)

#librosa.output.write_wav('testsongs/example1.wav', x, sr)


print(type(x), type(sr))
print(x.shape, sr)


plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)

X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()



x, fs = librosa.load(path_of_audio)
librosa.display.waveplot(x, sr=sr)

hop_length = 512
chromagram = librosa.feature.chroma_stft(x, sr=sr, hop_length=hop_length)
plt.figure(figsize=(15, 5))
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')




ipd.Audio(path_of_audio)







```
